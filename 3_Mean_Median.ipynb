```python
# =========================================================
# 3) Mean and Median Imputation (numeric) — Python
#     - Creates df_mean_imp and df_median_imp

#     - Preserves raw columns; adds *_imp columns
# =========================================================
import pandas as pd
from pathlib import Path

# 0) Load data (adjust path to your environment)
csv_path = Path("C:/Users/customer/Desktop/MyVideos/UDEMY LESSONS/IFRS 9 Python/DQ/with_missing_ifrs9_pit_pd.csv")  # e.g., r"C:\...\with_missing_ifrs9_pit_pd.csv"
df = pd.read_csv(csv_path)

# Numeric variables (from SAS example)
num_cols = ["age", "credit_utilization", "internal_score"]
num_cols = [c for c in num_cols if c in df.columns]

# ---------- Mean version ----------
df_mean_imp = df.copy()
for c in num_cols:
    mean_val = df_mean_imp[c].mean(skipna=True)
    df_mean_imp[f"{c}"]          = df_mean_imp[c]            # keep raw (explicit for clarity)
    df_mean_imp[f"{c}_imp_mean"] = df_mean_imp[c].fillna(mean_val)
```

```python

# PROC MEANS equivalent: n, nmiss, mean, median
summary_mean = []
for c in num_cols:
    summary_mean.append({
        "variable": c,
        "n": int(df_mean_imp[c].shape[0]),
        "nmiss": int(df_mean_imp[c].isna().sum()),
        "mean_after": float(df_mean_imp[f"{c}_imp_mean"].mean()),
        "median_after": float(df_mean_imp[f"{c}_imp_mean"].median()),
    })
summary_mean = pd.DataFrame(summary_mean)
print("=== Mean imputation summary ===")
print(summary_mean)

print("\n=== Mean-imputed sample (first 30 rows) ===")
df_mean_display = df_mean_imp[[c for c in df_mean_imp.columns if c in num_cols or c.endswith("_imp_mean")]]
df_mean_display.head(30)
```

```python
# ---------- Median version ----------
df_median_imp = df.copy()

for c in num_cols:
    med_val = df_median_imp[c].median(skipna=True)
    df_median_imp[f"{c}_imp_median"] = df_median_imp[c].fillna(med_val)

# ✅ Create a full dataframe containing raw + median-imputed versions
df_median_display = df_median_imp[
    [c for c in df_median_imp.columns if c in num_cols or c.endswith("_imp_median")]
].head(10)

# ✅ Show the table
df_median_display.head()
```

```python
# PROC MEANS equivalent: n, nmiss, mean, median
summary_median = []
for c in num_cols:
    summary_median.append({
        "variable": c,
        "n": int(df_median_imp[c].shape[0]),
        "nmiss": int(df_median_imp[c].isna().sum()),
        "mean_after": float(df_median_imp[f"{c}_imp_median"].mean()),
        "median_after": float(df_median_imp[f"{c}_imp_median"].median()),
    })
summary_median = pd.DataFrame(summary_median)
print("\n=== Median imputation summary ===")

print("\n=== Median-imputed sample (first 30 rows) ===")
print(df_median_imp[[c for c in df_median_imp.columns if c in num_cols or c.endswith("_imp_median")]].head(30))

# Optional: save outputs
df_mean_imp.to_csv("mean_imp_python.csv", index=False)
df_median_imp.to_csv("median_imp_python.csv", index=False)
summary_median.head()
```
