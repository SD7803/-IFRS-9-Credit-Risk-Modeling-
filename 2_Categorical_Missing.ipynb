```python
# =========================================================
# 0) Import demo dataset with artificial missing values
# =========================================================
import pandas as pd

# File path (adjust if needed)
csv_path = "C:/Users/customer/Desktop/MyVideos/UDEMY LESSONS/IFRS 9 Python/DQ/with_missing_ifrs9_pit_pd.csv"

# Load CSV (similar to PROC IMPORT)
df = pd.read_csv(csv_path)

print("✅ Dataset Loaded")
print("Shape:", df.shape)
print(df.head())

print("\n=== Data Types (PROC CONTENTS equivalent) ===")
print(df.dtypes)

```

```python
df.head(30)
```

```python

# =========================================================
# 1) Missing Level for Categorical (treat missing as 'Missing')
# =========================================================

# Identify categorical columns automatically
cat_cols = [c for c in df.columns if df[c].dtype == "object"]

print("\nCategorical columns detected:", cat_cols)

# Replace missing with 'Missing' (SAS: coalescec())
df_miss = df.copy()
for c in cat_cols:
    df_miss[c] = df_miss[c].fillna("Missing").astype(str).replace({"": "Missing"})


# =========================================================
# 2) Frequency tables (PROC FREQ equivalent)
# =========================================================
print("\n=== Frequency Tables (PROC FREQ equivalent) ===")
for c in cat_cols:
    print(f"\n--- {c} ---")
    print(df_miss[c].value_counts(dropna=False))


# =========================================================
# 3) Print first 200 rows (PROC PRINT OBS=200)
# =========================================================
print("\n=== First 200 Rows (PROC PRINT OBS=200) ===")
print(df_miss.head(10))


# Save cleaned version for next steps
output_path = "/mnt/data/demo_missing_level_python.csv"
df_miss.to_csv(output_path, index=False)

print("\n✅ Cleaned dataset saved to:", output_path)
```
